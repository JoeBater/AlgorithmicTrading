# gold prices

use the yahoo finance API to get gold prices

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

import datetime as dt
from datetime import date
from datetime import timedelta
```

```python
from pandas_datareader import data as pdr
import yfinance as yfin
yfin.pdr_override()
```

```python
start = '2021-01-01'
end = date.today() - timedelta(days = 1)
gold = pdr.get_data_yahoo('GC=F',start,end)
#silver = yf.download('SI=F',start,end)
#platinum = yf.download('PLG',start,end)

#gold = pd.read_csv(r'C:\Users\joeba\OneDrive\Documents\ML\gold prices\gold.csv')
```

```python
price = gold.tail(30).copy()
price
```

```python
plt.figure(figsize=(16,6))

#mpf.plot(gold,type='candle',style='yahoo')

width = 0.4
width2 = 0.01

plt.plot(price['Close'], color='black', linestyle='--', alpha=0.5 )

up = price[price['Close'] >= price['Open']]
dn = price[price['Close'] < price['Open']]

rd = 'red'
gn = 'green'

# plot up prices
plt.bar(up.index, up.Close-up.Open, width, bottom=up.Open, color='green')
plt.bar(up.index, up.High-up.Close, width2, bottom=up.Open, color='green')
plt.bar(up.index, up.Low-up.Open, width2, bottom=up.Open, color='green')

# plot dn prices
plt.bar(dn.index, dn.Close-dn.Open, width, bottom=dn.Open, color='red')
plt.bar(dn.index, dn.High-dn.Close, width2, bottom=dn.Open, color='red')
plt.bar(dn.index, dn.Low-dn.Open, width2, bottom=dn.Open, color='red')

fst = price.index[0].strftime('%Y-%m-%d')
lst = price.index[-1].strftime('%Y-%m-%d')

plt.xticks(pd.date_range(fst, lst).tolist(), rotation='vertical')

plt.show()
```
